```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix

df = pd.read_csv("D:/ML/Logistic/weather_forecast_data_Classification_Regression.csv")

encoder = OrdinalEncoder()
df["Rain"] = encoder.fit_transform(df[["Rain"]])

x = df.iloc[:,:-1]
y = df.iloc[:,-1]

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2, random_state = 42)


rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(x_train,y_train)

y_pred = rf_model.predict(x_test)

print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred, target_names=["No Rain", "Rain"]))
```

```python
train_under=x_train.copy()
train_under["rain"]=y_train
```

```python
df_class_0 = train_under[train_under['rain']==0]
df_class_1 = train_under[train_under['rain']==1]
```

```python
count_class_0, count_class_1 = train_under.rain.value_counts()
df_class_0 = df_class_0.sample(count_class_1)
```

```python
df_new=pd.concat([df_class_0,df_class_1],axis = 0)
```

```python
x_train_balanced =df_new.iloc[:,:-1]
y_train_balanced = df_new.iloc[:,-1]
```

```python
model_under = RandomForestClassifier(n_estimators = 100, random_state = 42)
model_under.fit(x_train_balanced, y_train_balanced)
```

```python
y_pred_balanced = model_under.predict(x_test)
```
