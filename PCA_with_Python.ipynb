```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
data = {
    'Height_cm': [170, 180, 160, 165, 175, 185, 155, 178, 168, 182],
    'Weight_kg': [65, 85, 50, 55, 75, 90, 45, 80, 60, 88],
    'Shoe_size': [8, 11, 6, 7, 9.5, 12, 5.5, 10.5, 7.5, 11.5]
}
df = pd.DataFrame(data)
print(df)
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)

# Optionally convert back to DataFrame
scaled_df = pd.DataFrame(scaled_data, columns=df.columns)
print(scaled_df.head(10))
```

```python
# Reduce to 2 principal components for visualization
pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_data)

# Create a DataFrame with the results
pca_df = pd.DataFrame(pca_result, columns=['PC1', 'PC2'])
print(pca_df.head(10))
```

```python
print("Explained Variance Ratio:", pca.explained_variance_ratio_)
print("Total Variance Explained:", sum(pca.explained_variance_ratio_))
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(pca_df['PC1'], pca_df['PC2'], c='blue', edgecolor='k')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Height, Weight, and Shoe Size')
plt.grid(True)
plt.show()
```
